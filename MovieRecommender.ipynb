```python
movies = movies[['movie_id', 'title', 'genres', 'keywords', 'overview',  'cast',
       'crew', 'vote_average', "release_date"]]
```

```python
movies.head()
```

```python
import ast
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def convertDir(text):
    L = []
    for i in ast.literal_eval(text):
        if i["job"] == "Director":
            L.append(i['name'])
    return L 
```

```python
movies["crew"] = movies["crew"].apply(convertDir)
```

```python
movies.dropna(inplace = True)
movies.isnull().sum()
```

```python
movies["overview"] = movies["overview"].apply(lambda x:x.split())
```

```python
movies.rename({"vote_average":"rating", "release_date":"year"}, axis =1, inplace = True)
```

```python
movies["year"] = movies["year"].apply(lambda x: x[:4])
```

```python
movies["tags"] = movies["genres"] + movies["keywords"] + movies["overview"] + movies["cast"] + movies["crew"]    
```

```python
movies.tags = movies["tags"].apply(lambda x:"".join(x))
```

```python
movies = movies[["movie_id", "title", "tags", "rating", "year"]]
movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
vector = cv.fit_transform(movies['tags']).toarray()
vector.shape
# type(vector)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
# new[new['title'] == 'The Lego Movie'].index[0]
```

```python
type(similarity)
```

```python
import sqlite3
```

```python
chunk_size = 1000
num_chunks = (similarity.shape[0] + chunk_size - 1) // chunk_size

for chunk_idx in range(num_chunks):
    start_idx = chunk_idx * chunk_size
    end_idx = min((chunk_idx + 1) * chunk_size, similarity.shape[0])
    
    # Connect to the SQLite database for the current chunk
    conn = sqlite3.connect(f'similarity_chunk_{chunk_idx}.db')
    c = conn.cursor()

    # Create the table schema
    c.execute('''CREATE TABLE IF NOT EXISTS similarity
                 (row_id INTEGER PRIMARY KEY, data BLOB)''')

    # Insert the data from the current chunk into the table
    for index, row in enumerate(similarity[start_idx:end_idx]):
        row_bytes = row.tobytes()
        c.execute("INSERT INTO similarity (row_id, data) VALUES (?, ?)", (index, row_bytes))

    # Commit the changes and close the connection
    conn.commit()
    conn.close()
```

```python
conn = sqlite3.connect('similarity.db')
c = conn.cursor()

c.execute('''CREATE TABLE IF NOT EXISTS similarity
             (row_id INTEGER PRIMARY KEY, data BLOB)''')
```

```python
for index, row in enumerate(similarity):
    row_bytes = row.tobytes()
    c.execute("INSERT INTO similarity (row_id, data) VALUES (?, ?)", (index, row_bytes))

conn.commit()
```

```python
c.execute("SELECT * FROM similarity")
rows = c.fetchall()

retrieved_similarity = []

for row in rows:
    row_id, data = row
    numpy_row = np.frombuffer(data, dtype=float)
    retrieved_similarity.append(numpy_row)

retrieved_similarity = np.array(retrieved_similarity)
```

```python
def recommend(movie):
    index = movies[movies['title'].apply(lambda x:x.lower()) == movie.lower()].index[0]
    distances = sorted(list(enumerate(retrieved_similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(f"{movies.iloc[i[0]].title} ({movies.iloc[i[0]].year}, {movies.iloc[i[0]].rating}*)")
```

```python
recommend("Avatar")
```

```python
def split_pickle(file, split_size):
    with open(file, 'rb') as f:
        data = pickle.load(f)
    split_data = [data[i:i+split_size] for i in range(0, len(data), split_size)]
    for i, d in enumerate(split_data):
        with open(f"{file}_{i}.pkl", 'wb') as f:
            pickle.dump(d, f)

split_pickle("similarity.pkl", 1000)
```
